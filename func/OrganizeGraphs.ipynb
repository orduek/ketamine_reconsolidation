## Creating the graphics
- Here we use PIL to concatenate the amygdala and hippocampus reactivity graphs into one and annotate it (a and b)
- This method can be further used to maybe add the figure itself or anything else needed

```python
from PIL import Image, ImageDraw, ImageFont 
```

```python
im1 = Image.open('amygdalaReactivity.png')
im2 = Image.open('hippoReactivity.png')
```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGBA', (im1.width + im2.width + 100, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width + 100, 0))
    return dst
```

```python
get_concat_h(im1, im2).save('amg_hippoReactivity.png')
```

```python
image = Image.open('amg_hippoReactivity.png') 
  
draw = ImageDraw.Draw(image)  
  
# specified font size 

font = ImageFont.truetype("/usr/share/fonts/dejavu/DejaVuSans.ttf", 80)
  
# drawing text size 
draw.text((2, 2400), "(a)", fill ="black", font = font, align ="right")  
draw.text((2400, 2400), "(b)", fill ="black", font = font, align ="right")  
  
image.save('amg_hippoReactivity.png')
```
